### Question 2: Qu'est ce que SQLite 3 ? Avantages et/ou Inconvénients ?

SQLite est un système de base de données qui a la particularité de fonctionner sans serveur, on dit aussi "standalone" ou "base de données embarquée". On peut l'utiliser avec beaucoup de langages : PHP, Python, C# (.NET), Java, C/C++, Delphi, Ruby...

inconvénients : 
- écritures atomiques (avec SQLite, un seul processus peut écrire dans la base)
- pas de réplication

avantages : 
- très léger et rapide à mettre en place
- stocker des données dans une vraie base de données sur une application pour smartphone (iPhone ou Android), pour une application Windows, ou sur un serveur web.

# Définition
réplication : La réplication consiste en un transfert des données d'un serveur vers un ou plusieurs autres serveurs SQL.
